```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
```

```python
def dataset_extract (url): 
    page = requests.get(url)
    
    print(page.status_code)
    
    print(page.content)
    
    soup = BeautifulSoup(page.content, 'html.parser')
    print(soup.prettify())
    
    tb = soup.find('table')
    
    for row in tb.find_all('td'):
        name = row.find('span')
        print(name)
    data = []
    for row in tb.find_all('td'):
        name = row.find('span')
        if name is not None:
            str = name.get_text()
        print(str)
        data.append(str)
        
    type(data)
    type(name)
    print(data[11:])
    d = data[11:]
    print(d)
        
    dt = ['State/UT', '2001', '2003', '2005', '2009', '2011', '2013', '2015', '2017']
    print(dt)
    
    it = [iter(dt)] * 9
    tuples = []
    for l in zip(*it):
        print(l)
        tuples.append(l)
    it = [iter(d)] * 9
    for l in zip(*it):
        print(l)
        tuples.append(l)
    print(tuples)
    
    df = pd.DataFrame(tuples)
    df.head()
    
    df1 = df.rename(columns=df.iloc[0])
    df1.head()
    
    tree = df1.drop(df1.index[0])
    tree.head()
    
    tree.to_csv('tree_cover.csv', na_rep='Null', float_format='%.2f', index=False, encoding='utf-8')
    return
```

```python
dataset_extract("http://www.frienvis.nic.in/Database/Tree-Cover-in-Indian-States-and-UTs_2468.aspx")
```
