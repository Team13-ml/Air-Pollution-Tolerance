```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv
import requests
```

```python
f1=pd.read_csv('air22.csv',encoding='unicode_escape')
f1.dtypes
```

```python
f1['date'] = pd.to_datetime(f1['date'], errors='coerce')
f1['year']=f1['date'].dt.year
f1.head()
```

```python
f1.drop(['stn_code','agency','spm','sampling_date','type','location_monitoring_station','formula','rspm','location'],axis=1,inplace=True)
f1.head()
```

```python
option=['Delhi','Haryana','Maharashtra','Kerala','Tamil Nadu','West Bengal','Gujarat']
f1=f1.loc[(f1['state'].isin(option)) & (f1['date'].dt.year>=2000)]

f1.head()
f1.fillna(0)
f1=f1.replace(np.nan,0)
```

```python
dd=f1.groupby(['year','state']).agg('mean')
dd
```

```python
f2=pd.read_csv('tamilnadu.csv', encoding='utf-8')
f2['date'] = pd.to_datetime(f2['date'], errors='coerce')
f2['year']=f2['date'].dt.year
f2.head()
```

```python
df1=f2.drop([' pm10',' o3',' co'],axis=1)
df1.fillna(0.0)
```

```python
for i in df1['pm2_5']:
    df1['pm2_5']=float(i)
for j in df1['no2']:
    df1['no2']=float(j)
for k  in df1['so2']:
    df1['so2']=float(k)
for l in df1['AQI']:
    df1['AQI']=float(l)
df1.dtypes
```

```python
df2=df1.groupby(['year','state']).agg('mean')
df2.head()
```

```python
aa=pd.concat([dd,df2],sort=False,axis=0)
y1=aa.to_csv('newr.csv')
```
