```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.frienvis.nic.in/Database/Tree-Cover-in-Indian-States-and-UTs_2468.aspx"
page = requests.get(url)
```

```python
print(page.status_code)
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
tb = soup.find('table')
```

```python
for row in tb.find_all('td'):
    name = row.find('span')
    print(name)
```

```python
data = []
for row in tb.find_all('td'):
    name = row.find('span')
    if name is not None:
        str = name.get_text()
    print(str)
    data.append(str)
```

```python
dt = ['State/UT', '2001', '2003', '2005', '2009', '2011', '2013', '2015', '2017', '2019']
print(dt)
```

```python
it = [iter(dt)] * 10
```

```python
tuples = []
for l in zip(*it):
    print(l)
    tuples.append(l)
```

```python
it = [iter(d)] * 9
```

```python
df1 = df.rename(columns=df.iloc[0])
df1.head()
```
